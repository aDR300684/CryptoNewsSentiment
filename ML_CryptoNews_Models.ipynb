To improve the performance of the Random Forest classifier, we will now conduct hyperparameter tuning using GridSearchCV. This method will exhaustively search over the specified parameter grid to find the combination of parameters that yields the best cross-validated accuracy.

```python
# Random Forest Hyperparameter Tuning with GridSearchCV
from sklearn.model_selection import GridSearchCV

# This is the third iteration of hyperparameter tuning using GridSearchCV.
# The parameter grid below has been refined based on results from the previous two rounds.
# We are focusing on a narrower range of values around the best parameters found so far,
# aiming for fine-tuning to improve the model's performance incrementally.
param_grid = {
    'n_estimators': [120, 130, 140],  # Narrow down around the best found
    'max_depth': [15, 16, 17],  # Explore slightly around the best max_depth
    'min_samples_split': [11, 12, 13],  # Fine-tune around the best value
    'min_samples_leaf': [1, 2, 3],  # The best was 2
    'bootstrap': [True]  # Keeping as fixed since it was found beneficial
}


# Initialize the Random Forest Classifier
random_forest = RandomForestClassifier(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=random_forest, param_grid=param_grid, 
                           cv=5, n_jobs=-1, verbose=10, scoring='accuracy')

# Perform the search on the training set
grid_search.fit(X_train_new, y_train_new)

# Output the best parameters and the best score from the search
best_parameters = grid_search.best_params_
best_score = grid_search.best_score_

print(f"Best parameters found: {best_parameters}")
print(f"Best cross-validated score found: {best_score}")
```

### Transition to Bayesian Optimization with Optuna

Given the diminishing returns from further exhaustive grid searches, we will next explore Bayesian optimization using Optuna for hyperparameter tuning. Bayesian optimization is a more efficient strategy for hyperparameter optimization, leveraging past trial results to inform the selection of the next set of parameters to evaluate. This approach is particularly suited for scenarios where we seek to minimize the number of evaluations due to the computational expense of each model training and validation process.

Optuna provides a flexible framework for defining a search space and automates the process of selecting the most promising hyperparameters to evaluate next, based on the performance of previous trials. By using Optuna, we aim to explore the hyperparameter space more effectively, potentially uncovering a combination of parameters that could yield better performance than what has been achieved so far with `GridSearchCV`.

```python
import optuna
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
def objective(trial):
    # Define the hyperparameters to tune via Optuna
    n_estimators = trial.suggest_int('n_estimators', 100, 200)
    max_depth = trial.suggest_int('max_depth', 10, 30)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 14)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 4)
    bootstrap = trial.suggest_categorical('bootstrap', [True, False])
    
    # Initialize the model with the hyperparameters
    clf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth,
                                 min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf,
                                 bootstrap=bootstrap, random_state=42, verbose=10)
    
    # Perform cross-validation and return the mean accuracy
    scores = cross_val_score(clf, X_train_new, y_train_new, cv=5, scoring='accuracy')
    return np.mean(scores)
```

```python
study = optuna.create_study(direction='maximize', study_name='RF Optimization')
study.optimize(objective, n_trials=50, show_progress_bar=True)

# After the optimization, print the best parameters and the best score
print(f"Best parameters: {study.best_params}")
print(f"Best cross-validated score: {study.best_value}")
```

### Interpretation of Optuna Optimization Results

Following an extensive exploration of the hyperparameter space using Optuna's Bayesian optimization framework, we arrived at a set of parameters that produced the best cross-validated accuracy score of approximately 0.6036 for our Random Forest model. The best-performing hyperparameters identified through this process are as follows:

- `bootstrap`: False
- `max_depth`: 14
- `min_samples_leaf`: 4
- `min_samples_split`: 10
- `n_estimators`: 153

These parameters were determined after assessing various combinations through a series of trials, where the model's performance was closely monitored and optimized. Despite the high granularity of the hyperparameter search, the resulting best cross-validated score from Optuna was marginally lower than what we achieved with `GridSearchCV`, suggesting that the potential for improvement within the Random Forest algorithm is becoming increasingly marginal with our current dataset and feature configuration.

The convergence of scores from both `GridSearchCV` and Optuna around the ~0.60 mark may indicate that we are reaching the inherent limits of what the Random Forest model can discern from the data, given the existing features and model complexity. While Optuna provided an efficient and intelligent search strategy, the results imply that alternative approaches or models might be necessary to achieve substantial performance gains beyond this point.

### Hyperparameter Tuning Setup for SVM

Before running the hyperparameter tuning process, it's crucial to define the range of parameters to search. For the SVM model, the most critical hyperparameters include the regularization penalty `C`, the `gamma` value that influences the decision boundary, and the choice of `kernel`.

The regularization parameter `C` controls the trade-off between achieving a low training error and a low testing error — that is, between overfitting and underfitting the model. A lower value of `C` creates a small-margin hyperplane and a more regularized model, while a higher value of `C` aims for a larger-margin hyperplane and a less regularized model.

`gamma` is a parameter for non-linear kernels to handle the trade-off between the decision boundary's complexity and the amount of variance the model allows for. When `gamma` is low, the model's decision boundary is smoother, while a high `gamma` leads to a more complex, tightly-fitting decision boundary.

We employ `GridSearchCV` to conduct an exhaustive search over specified parameter values for our SVM model. The `StandardScaler` within a pipeline ensures that feature scaling is applied to the data before model fitting, which is essential for optimal SVM performance. The grid search will use 5-fold cross-validation to assess the model's performance, thus providing a robust evaluation scheme.


```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Define the hyperparameter space for the grid search
# Note: A comprehensive and exhaustive grid search was previously conducted with 480 fits to identify 
# optimal hyperparameters for the SVM model. This process involved a wider range of values 
# and was computationally intensive, requiring significant processing time.
# For the purposes of this demonstration and to ensure a more time-efficient execution, 
# we are narrowing the hyperparameter space based on insights gained from the full search.
# This allows us to focus on a subset of hyperparameters that were identified as promising, 
# thus optimizing computational resources while still aiming for high model performance. 
param_grid = {
    'svc__C': [1, 10, 100],  # Explore around the optimal C value
    'svc__gamma': [0.0001, 0.001, 0.01],  # Narrowed focus around the optimal gamma
    'svc__kernel': ['rbf'],  # Keeping the kernel fixed at 'rbf' as identified optimal
}

# Create a pipeline that standardizes the data then applies SVM
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', SVC())
])

# Initialize the GridSearchCV object
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', verbose=10, n_jobs=-1)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# The best hyperparameters from the grid search
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best cross-validated score: {grid_search.best_score_}")
```

## Understanding GBM and Its Relevance to Our Project

Gradient Boosting Machines (GBM) are powerful machine learning algorithms known for their effectiveness in classification tasks across a wide range of domains. GBM works by building an ensemble of weak predictive models, typically decision trees, in a sequential manner where each subsequent model attempts to correct the errors of its predecessor. This process continues until the addition of new models does not significantly improve the performance, or a specified number of trees is reached.

For our project, which involves predicting price momentum trends based on cryptocurrency news, GBM could be particularly relevant due to its ability to capture complex non-linear patterns and interactions between features. Our dataset, characterized by various sentiment scores and topic weights derived from textual analysis, presents a challenge well-suited to GBM's strengths.
```python
import joblib

# Define the path to the split datasets
split_train_val_dataset_path = r"C:\Users\adrco\Final_Project-env\Datasets\Splits\Train_Val"

# Filenames for the newly split datasets
split_train_val_filenames = ['X_train_new.pkl', 'X_val.pkl', 'y_train_new.pkl', 'y_val.pkl']

# Load the datasets
X_train_new = joblib.load(f"{split_train_val_dataset_path}\\X_train_new.pkl")
X_val = joblib.load(f"{split_train_val_dataset_path}\\X_val.pkl")
y_train_new = joblib.load(f"{split_train_val_dataset_path}\\y_train_new.pkl")
y_val = joblib.load(f"{split_train_val_dataset_path}\\y_val.pkl")
```

```python
import optuna
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

def objective(trial):
    # Hyperparameter ranges
    n_estimators = trial.suggest_int('n_estimators', 50, 500)
    learning_rate = trial.suggest_float('learning_rate', 0.01, 0.3, log=True)  
    max_depth = trial.suggest_int('max_depth', 3, 10)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 20)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 10)
    
    # Model setup
    gbm = GradientBoostingClassifier(
        n_estimators=n_estimators, 
        learning_rate=learning_rate,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        random_state=42
    )
    
    # Cross-validation
    scores = cross_val_score(gbm, X_train_new, y_train_new, cv=3, scoring='accuracy', n_jobs=-1)  

    return np.mean(scores)

# Optuna optimization
study = optuna.create_study(direction='maximize', study_name='GBM Optimization')
study.optimize(objective, n_trials=10, show_progress_bar=True)  # Reduced trials 

print("Number of finished trials: ", len(study.trials))
print("Best trial:")
trial = study.best_trial

print("  Value: ", trial.value)
print("  Params: ")
for key, value in trial.params.items():
    print(f"    {key}: {value}")

```

### Introduction to Random Search for Hyperparameter Tuning in Deep Neural Networks

Random Search is an efficient approach for hyperparameter tuning, especially when dealing with complex models like Deep Neural Networks (DNN). Unlike grid search, which systematically evaluates all possible combinations of hyperparameters, Random Search randomly samples the hyperparameter space and evaluates a fixed number of parameter settings. This stochastic nature of Random Search often leads to finding a good enough set of hyperparameters more quickly and with fewer computations, making it particularly well-suited for DNNs, which can have an extensive range of hyperparameters.

The advantages of Random Search include:
- Efficiency: It can stumble upon an optimal or near-optimal solution without the need to explore the entire space exhaustively.
- Flexibility: It can easily handle discrete, continuous, and conditional dimensions.
- Scalability: It is less affected by the dimensionality of the hyperparameter space.

By randomly sampling the space, Random Search can explore a more diverse set of values and often yields better results in a shorter time, especially when the optimal hyperparameters are located in a narrow region that grid search might miss.


```python
import tensorflow as tf
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
from scikeras.wrappers import KerasClassifier

# Function to create the DNN model required by KerasClassifier
def create_model(neurons=64, activation='relu', optimizer='adam', dropout_rate=0.5):
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(neurons, input_dim=X_train_scaled.shape[1], activation=activation),
        tf.keras.layers.Dropout(dropout_rate),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# Adjusted param_dist without 'epochs' and 'batch_size'
param_dist = {
    'model__neurons': [64, 128, 256],
    'model__activation': ['relu', 'tanh'],
    'model__optimizer': ['adam', 'rmsprop'],
    'model__dropout_rate': [0.0, 0.2, 0.5],
}

# Wrap the model with KerasClassifier, specifying epochs and batch_size here
model = KerasClassifier(model=create_model, epochs=200, batch_size=32, verbose=0)

# Create and fit RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=10, n_jobs=-1, cv=5, verbose=10)
random_search_results = random_search.fit(X_train_scaled, y_train_new)

# Summarize results
print("Best: %f using %s" % (random_search_results.best_score_, random_search_results.best_params_))

```

By employing these techniques, we aim to improve the model's ability to generalize and thus enhance its performance on unseen data. This will be a crucial step towards developing a reliable and robust ML pipeline.
